## Imports

```python
import pandas as pd
import numpy as np

from scipy.interpolate import LSQUnivariateSpline
import matplotlib.pyplot as plt

from skfda.representation.basis import BSpline
from skfda.representation.basis import BSplineBasis

from skfda.representation import FDataGrid
from sklearn.metrics import mean_squared_error

from scipy.interpolate import UnivariateSpline
from sklearn.model_selection import LeaveOneOut
from sklearn.metrics import mean_squared_error

from sklearn.kernel_ridge import KernelRidge
```

## Read in Data

```python
data = pd.read_csv('coal_energy_data.csv', header=None)
data.columns = ['year', 'energy']
xs = np.array(data['year'])
ys = np.array(data['energy'])
data[:5]
```

### Cubic Spline
----------------------
Calculate best number of knots and associated MSE

Plot the cubic spline model with the optimal number of knots

```python
def cubic_spline_mse(x, y, n_knots):
    mse_values = []  # list to hold the MSE values
    
    for i in range(len(x)):
        x_train = np.delete(x, i) #LLO CV 
        y_train = np.delete(y, i) #LLO CV
        x_val = x[i] #validation point
        y_val = y[i] #validation point

        knots = np.linspace(x_train.min(), x_train.max(), n_knots+2)[1:-1] #interior knots without endpoints

        spline = LSQUnivariateSpline(x_train, y_train, knots) #fit a LSQ Univariate Spline

        y_pred = spline(x_val) #prediction for validation point
        
        se = (y_val - y_pred) ** 2 #squared error
        mse_values.append(se)
    
    mse = np.mean(mse_values)
    return mse

for i in range(6,16):
    mse = cubic_spline_mse(xs,ys,i)
    print("knots:", i, "\tcubic spline MSE:", mse)
```

```python
def cubic_spline_plot(x, y, n_knots):
    knots = np.linspace(x.min(), x.max(), n_knots+2)[1:-1] #interior knots without endpoints of x
    
    spline = LSQUnivariateSpline(x, y, knots) #fit LSQ Univerariate spline
    
    xnew = np.linspace(x[0], x[-1], 1000) #generate new values for creating line chart for spline
    
    ynew = spline(xnew) #generate new ys from cubic spline
    
    plt.scatter(x, y, label='Original points', color='black', marker='o', facecolors='none') #scatter original points
    
    plt.plot(xnew, ynew, label='Cubic Spline', color='red') #make new spline a line chart
    
    plt.title('Cubic Spline Model of Electrical Energy Produced by Coal in the US by Year')
    plt.xlabel('year')
    plt.ylabel('electrical energy units')
    plt.legend()
    plt.show()

cubic_spline_plot(xs, ys, 7)
```

### B-Spline MSE
-------------------
Calculate best number of knots and associated MSE

Plot the b-spline model with the optimal number of knots

```python
def bspline_mse(x, y, n_knots):
    domain_range = (np.min(x), np.max(x))
    order = 4  # For cubic splines, order is degree + 1

    # Adjust knots and n_basis appropriately
    n_internal_knots = n_knots - 2  # Remove boundary knots
    knots = np.linspace(domain_range[0], domain_range[1], n_knots)
    
    # Set n_basis based on n_knots and order, to ensure that we have a valid configuration
    n_basis = len(knots) + order - 2  

    errors = []

    for i in range(len(x)):
        # Create LOOCV training and test sets
        x_train = np.delete(x, i)
        y_train = np.delete(y, i)
        x_test = x[i]
        y_test = y[i]

        # Configure the basis with the correct number of basis functions and knots
        basis = BSplineBasis(domain_range=domain_range, n_basis=n_basis, order=order, knots=knots)

        fd_train = FDataGrid(data_matrix=y_train[None, :], grid_points=[x_train])
        fd_train_basis = fd_train.to_basis(basis)

        # Evaluate spline at the test point
        y_pred = fd_train_basis.evaluate([x_test]).squeeze()

        # Calculate and collect the squared error
        errors.append((y_test - y_pred) ** 2)

    # Compute the MSE from collected errors
    mse = np.mean(errors)
    return mse

for i in range(6,16):
    mse = bspline_mse(xs,ys,i)
    print("knots:", i, "\tBspline MSE:", mse)
```

```python
def bspline_plot(x, y, n_knots):
    plt.scatter(x, y, label='Original points', color='black', marker='o', facecolors='none')
    knots = np.linspace(np.min(x), np.max(x), n_knots)
    basis = BSpline(knots=knots, order=3)
    
    fd = FDataGrid(data_matrix=np.array(y)[None, :], grid_points=[x]) #fit b-spline to data
    fd_basis = fd.to_basis(basis)
    y_pred = fd_basis(x).squeeze()
    
    plt.plot(x, y_pred, color='red', label='B-spline')
    
    plt.xlabel('year')
    plt.ylabel('electrical energy units')
    plt.title('B-spline Model of Electrical Energy Produced by Coal in the US by Year')
    plt.legend()
    plt.show()
    
bspline_plot(xs, ys, n_knots=9)
```

### Smoothing Spline
-------------------------
Calculate best smoothing constant using Leave-One-Out Cross-Validation

Plot the smoothing spline model with the optimal smoothing constant $s$

```python
def smoothing_spline_mse(x, y):
    s_values = np.logspace(-1, 3, 5) #array of smoothing constant factors logarithmically spaced from 0.1 to 1000

    best_mse = float('inf')
    best_s = None
    loo = LeaveOneOut()
    
    for s in s_values: #loop over smooothing constant values
        mse_values = []
        for train_index, test_index in loo.split(x):
            x_train, x_test = x[train_index], x[test_index] #train test split with LOO CV
            y_train, y_test = y[train_index], y[test_index]
            
            spline = UnivariateSpline(x_train, y_train, s=s) #fit smoothing spline on training data
            
            y_pred = spline(x_test) #predict response for test data point
            
            mse_values.append(mean_squared_error(y_test, y_pred)) #MSE for current fold

        mse_avg = np.mean(mse_values) #avg MSE for current smoothing factor
        if mse_avg < best_mse:
            best_mse = mse_avg
            best_s = s
            
    return (best_mse, best_s)

smoothing_spline_vals = smoothing_spline_mse(xs, ys)
print("best smoothing lambda:", smoothing_spline_vals[1], "\tSmoothing spline MSE:", smoothing_spline_vals[0])
```

```python
def smoothing_spline_plot(xs, ys, s):
    plt.scatter(xs, ys, label='Original points', color='black', marker='o', facecolors='none')

    spline_model = UnivariateSpline(xs, ys, s=s) #generate smoothing spline model

    x_smooth = np.linspace(min(xs), max(xs), 1000) #generate 1000 points for model line chart
    y_smooth = spline_model(x_smooth)

    plt.plot(x_smooth, y_smooth, color='red', label='Smoothing Spline') #line chart creation

    plt.xlabel('year')
    plt.ylabel('electrical energy units')
    plt.title('B-spline Model of Electrical Energy Produced by Coal in the US by Year')
    plt.legend()
    plt.show()
    
smoothing_spline_plot(xs,ys,s=10)
```

### Kernel Regression with Gaussian Kernel
-------------------------
Calculate best Gaussian Kernel $\lambda$ value using Leave-One-Out Cross-Validation

Plot the Kernel Regression model with the optimal $\lambda$ value.

```python
def kernel_reg_mse(x, y):
    x = np.array(x).reshape(-1, 1) #change from (69,) to (69,1)
    y = np.array(y).reshape(-1, 1)

    lambdas = np.logspace(start=np.log10(.00001), stop=np.log10(1), num=50) #logarithmically search for lambda vals

    best_mse = float('inf')
    best_lambda = None
    loo = LeaveOneOut() #leave-one-out CV

    # Grid search over possible lambdas
    for lamb in lambdas:
        mse_values = []
        for train_index, test_index in loo.split(x): #LOO CV
            x_train, x_test = x[train_index], x[test_index]
            y_train, y_test = y[train_index], y[test_index]

            model = KernelRidge(kernel='rbf', gamma=lamb) #fit kernel regression model
            model.fit(x_train, y_train)

            y_pred = model.predict(x_test) #predict response for test point

            mse_values.append(mean_squared_error(y_test, y_pred)) #compute MSE of current LOO fold

        mse_avg = np.mean(mse_values) #avg MSE for current lambda value
        if mse_avg < best_mse: #if current MSE < best MSE
            best_mse = mse_avg
            best_lambda = lamb

    return (best_mse, best_lambda)

kernel_reg_mse(xs,ys)
kernel_reg_vals = kernel_reg_mse(xs, ys)
print("best kernel regression lambda:", kernel_reg_vals[1], "\tKernel regression MSE:", kernel_reg_vals[0])
```

```python
def kernel_regression_plot(xs, ys, lamb):
    plt.scatter(xs, ys, label='Original Points', color='black', marker='o', facecolors='none')

    kernel_model = KernelRidge(kernel='rbf', gamma=lamb) #generate kernel regression model
    kernel_model.fit(np.array(xs).reshape(-1, 1), ys) #fit model

    x_smooth = np.linspace(min(xs), max(xs), 1000).reshape(-1, 1) #generate line chart points
    y_smooth = kernel_model.predict(x_smooth)

    plt.plot(x_smooth, y_smooth, color='red', label='Kernel Regression') #make line chart

    plt.xlabel('Year')
    plt.ylabel('Electrical Energy Units')
    plt.title('Kernel Regression Model of Electrical Energy Produced by Coal in the US by Year')
    plt.legend()
    plt.show()
    
kernel_regression_plot(xs,ys,lamb=.00281177)
```
